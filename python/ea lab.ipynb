```python
df['is_train'] = np.random.uniform(0, 1, len(df)) <= 0.75
df_train = df.loc[df['is_train'] == True]
df_test = df.loc[df['is_train'] != True]
```

```python
f_train = df_train.iloc[:, 1:9]
f_test = df_test.iloc[:, 1:9]
```

```python
X_train = f_train.drop('Y house price of unit area', axis = 1)
y_train = f_train['Y house price of unit area']
X_test = f_test.drop('Y house price of unit area', axis = 1)
y_test = f_test['Y house price of unit area']
```

```python
model = LinearRegression() 
model.fit(X_train, y_train) 
pd.DataFrame(model.coef_, index = X_train.columns, columns = ['Coef'])
```

```python
y_pred = model.predict(X_test)
MAE = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test, y_pred)
RMSE = MSE ** (1/2)

pd.DataFrame([MAE, MSE, RMSE], index = ['MAE', 'MSE', 'RMSE'], columns = ['Metrics'])
```
