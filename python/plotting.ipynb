Plotting using independent x and y arrays


```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,10) #produces array in the range of [a,b) equally spaced
y=x*x             #elementwise matrix multiplication also called Hadamard multiplication
plt.title('Plot using x and y array')
plt.xlabel('X entries')
plt.ylabel('Y entries')
plt.plot(x, y, color='blue')
plt.show()
```

```python
plt.scatter(x, y, color='red')
plt.show()
```

```python
''' other way is to create a figure and add axes and axis to them
    axes are basically the plot itself and axis is, well the tick marked straight lines 
    which indicate the abcessa and ordinates themselves
'''
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots()
ax.plot(x, x, label='linear') #linear
ax.plot(x, x*x, label='quadratic') #quadratic
ax.plot(x, x*x*x, label='cubic') #cubic
ax.set_title('simple plot')
ax.legend()
```
