```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
photo=io.imread("pepe.png")
print(photo.shape)
plt.imshow(photo)
```

```python
from skimage import io
import matplotlib.pyplot as plt
photo=io.imread("sad.png")
print(photo.shape)
plt.imshow(photo[::5,::5])
io.imsave("sadlow.png",photo[::5,::5])
```

```python
from skimage import io
import matplotlib.pyplot as plt
photo=io.imread("pepe.png")
print(photo.shape)
plt.imshow(photo[:,::-1])
io.imsave("peperev.png",photo[::-1,::-1])
```

```python
from skimage import io
import matplotlib.pyplot as plt
photo=io.imread("pepe.png")
print(photo.shape)
plt.imshow(photo[80:112,70:112])
io.imsave("pepehands.png",photo[80:112,70:112])
```
